```
import gensim
from gensim.models import word2vec
from gensim.models import KeyedVectors
from sklearn.metrics.pairwise import cosine_similarity
```

**Keyed Vectors** -  this object essentially contains the mapping between words and embedding. After training it can be used directly to query those embedding in various ways.

```
input1=['apple','mango','juice','banana',"melon"]
input2=['cricket','hockey','chess','dance','football']
input3=['india','germany','russia','france','paris']
```

```
!wget -c "https://s3.amazonaws.com/dl4j-distribution/GoogleNews-vectors-negative300.bin.gz"
```

```
word_vectors= KeyedVectors.load_word2vec_format("/content/GoogleNews-vectors-negative300.bin.gz",binary=True)
```

```
v_apple=word_vectors['apple']
v_mango= word_vectors['mango']
print(v_apple.shape,v_mango.shape,sep="\n")
```

```
print(cosine_similarity([v_apple],[v_mango]))  #It accept 2D list  
```

**FIND ODD ONE OUT**

```
import numpy as np
```

```
def odd_one_out(words):
  # Accepts the words of the string and return odd word
  #generate all the word embeding of the given list
  
  all_word_vectors=[word_vectors[w] for w in words]
  
  #print(len(all_word_vectors))
  #print(np.array(all_word_vectors).shape)
  
  avg_vector = np.mean(all_word_vectors,axis=0)

  #Ilterarte over and over and find similarity

  odd_one=None
  min_similarity= 1.0  #we can any value greater

  for w in words:
    sim = cosine_similarity([word_vectors[w]],[avg_vector])
    if sim < min_similarity:
      min_similarity =  sim
      odd_one= w
    
    print("similarity between {} and avg_vector is {}".format(w,sim))
      
  return odd_one
```

```
print(odd_one_out(input1))
print(odd_one_out(input2))
print(odd_one_out(input3))
```
